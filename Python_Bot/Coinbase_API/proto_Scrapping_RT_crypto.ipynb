## Get HTML of Coinmarket

```python
import os, sys
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

DRIVER_PATH = os.path.join(os.getcwd(),"chromedriver.exe")
options = Options()
options.headless = False
options.add_argument("--window-size=1920,1200")
driver = webdriver.Chrome(DRIVER_PATH, options=options)

```

### Extract all cryptocurencies available in Coinbase

```python
import time

NB_PAGE = 3
MAX_TRY = 10

t_start = time.time()
for page in range(1,NB_PAGE+1):
    driver.get(f"https://www.coinbase.com/fr/price/s/listed?page={page}")

    ctr_try = 0
    while True:
        ctr_try+=1
        grid = driver.find_elements_by_css_selector("tbody")
        if grid[0].text!='':
            print(grid[0].text)
            break
        print('Retry')
        if ctr_try>=MAX_TRY:
            raise AssertionError('Too many try to get crypto data')
print('Delta time: ',time.time()-t_start) # Around 7s with a bad connection
```

### Make the task running every minute
(cannot use asyncio in notebook)

```python
import os, sys
import asyncio
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

DRIVER_PATH = os.path.join(os.getcwd(),"chromedriver.exe")
options = Options()
options.headless = False
options.add_argument("--window-size=1920,1200")
driver = webdriver.Chrome(DRIVER_PATH, options=options)

def refresh_task_every(task, delta_t = 60):
    while True:
        t_sart = time.time()
        # Do task
        task()
        # Sleep current period of time
        t_sleep = delta_t-time.time()+t_sart
        if t_sleep<0:
            print(f'WARNING: Task takes more than {delta_t}s to execute')
        time.sleep(max(t_sleep,0))

NB_PAGE = 3
MAX_TRY = 10

def refresh_crypto_value():
    t_start = time.time()
    for page in range(1,NB_PAGE+1):
        driver.get(f"https://www.coinbase.com/fr/price/s/listed?page={page}")

        ctr_try = 0
        while True:
            ctr_try+=1
            grid = driver.find_elements_by_css_selector("tbody")
            if grid[0].text!='':
                print(grid[0].text)
                break
            print('Retry')
            if ctr_try>=MAX_TRY:
                raise AssertionError('Too many try to get crypto data')

refresh_task_every(refresh_crypto_value,delta_t=60)
```
