## Call Coinbase API

```python
import os
import json
# import coinbase
from datetime import date, datetime, timedelta
import urllib.parse
import urllib
import requests

MAX_REQUEST = 300

def date2coinbaseDate(t):
    return t.isoformat(timespec='seconds') + '+00:00'

date_end = datetime.now()
dt = timedelta(minutes=MAX_REQUEST)
date_start = date_end - dt

# Generate Coinase Url
params = {'start': date2coinbaseDate(date_start), 'end': date2coinbaseDate(date_end), 'granularity': 60}
crypto = 'ETH'
url = f"https://api.pro.coinbase.com/products/{crypto}-EUR/candles?\
{urllib.parse.urlencode(params)}"
url
# Request
r = requests.get(url)
succeed = str(r.status_code)[0]=='2'
if not succeed:
    raise ConnectionError(r.json()['message'])
data = r.json()
```

## Exctract lowest price

```python
# Use second line that corresponds to lowest price
data_serie = {}
for d in data:
    data_serie[d[0]] = d[1]
print(data_serie)
```
